# K-means Tutorial

K-means is an unsupervised machine learning technique that classifies data into K clusters based on their similarities. The "K" in K-means represents the number of clusters that the algorithm tries to identify and the "means" are the means of the clusters. 

During the clustering process, K-means clustering divides the data into K clusters, each with K centroids, and then calculates the means of each cluster. The algorithm aims to minimize the WCSS, which measures the sum of the squared distances between each data point and its assigned centroid. By minimizing the WCSS, K-means clustering tries to create compact and well-separated clusters. It starts with k random centroids. It adjusts the centroids based on the means and then iteratively updates the centroid positions until the centroids stop moving or a maximum number of iterations is reached.

K-means clustering assumes that the clusters have equal variances, which may not always be the case in real-world scenarios. If the variance within clusters is significantly different, the K-means algorithm may not perform well, and it may be necessary to use a different clustering algorithm that can handle clusters with varying variances.

```python
# Import the necessary packages
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

```

```python
# Let's generate a random dataset
np.random.seed(0)

# This will generate a list of 200 points where each x and y value
# are drawn from a normal distribution with mean 0 and variance 1
X = np.random.randn(200, 2)
```

```python
# Visualize the dataset using a scatter plot
plt.scatter(X[:,0], X[:,1])
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Perform k-means clustering with k=3 clusters
It will try to find similarities in this dataset

```python
kmeans = KMeans(n_clusters=3, random_state=42,)
kmeans.fit(X)
```

```python
# You can use the .labels_ method to see the classification labels
print(kmeans.labels_)
```

```python
# You can use the .cluster_centers_ method to see the coordinates of the centroids
print(kmeans.cluster_centers_)
```

```python
# You can use n_iter_ to check how many iterations it conducted
print(kmeans.n_iter_)
```

```python
# Store cluster labels and centroid coordinates
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Create scatterplot of dataset with color-coded clusters
plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='r')
plt.title('K-Means Clustering Results')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Fit k-means model with one iteration
kmeans1 = KMeans(n_clusters=3, random_state=42, max_iter=1)
kmeans1.fit(X)
labels1 = kmeans1.labels_
centroids1 = kmeans1.cluster_centers_
print(labels1)
print(centroids1)
print(kmeans1.n_iter_)
```
